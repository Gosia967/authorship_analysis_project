```python
import os
import sys
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
print(module_path)
print(sys.path)
```

```python
e_basic = dict()
l_basic = dict()
for dir in os.listdir():
  if dir.startswith('set_a_'):
    print(dir)
    print(os.listdir(dir))
    for author in os.listdir(dir):
      char_num = 0
      words_num = 0
      for file_name in os.listdir(os.path.join(dir, author)):
        with open(os.path.join(dir, author, file_name)) as file:
          text = file.read()
          char_num += len(text)
          words = text.split()
          words_num += len(words)
      if dir.startswith('set_a_e_'):
        e_basic[author] = (char_num, words_num)
      if dir.startswith('set_a_l_'):
        l_basic[author] = (char_num, words_num)
      print(f'{author}: number of characters: {char_num}, number of words: {words_num}')
```

```python
for dir in os.listdir():
  if dir.startswith('set_a_'):
    print(dir)
    print(os.listdir(dir))
    for author in os.listdir(dir):
      char_num = 0
      words_num = 0
      for file_name in os.listdir(os.path.join(dir, author)):
        with open(os.path.join(dir, author, file_name)) as file:
          text = file.read()
          char_num += len(text)
          words = text.split()
          words_num += len(words)
      if dir.startswith('set_a_e_'):
        words_str = str(round(words_num/(10**6), 1)).replace('.',',')
        chars_str = str(round(char_num/(10**6), 1)).replace('.',',')
        print(f'{author} & ${words_str}*10^6$ & ${chars_str}*10^6$ \\\\')
      if dir.startswith('set_a_l_'):
        words_str = str(round(words_num/(10**3), 1)).replace('.',',')
        chars_str = str(round(char_num/(10**3), 1)).replace('.',',')
        print(f'{author} & ${words_str}*10^3$ & ${chars_str}*10^3$ \\\\')
```

```python
sorted_dir = sorted(os.listdir(), key=len)
for dir in sorted_dir:
  if dir.startswith('set'):
    print(dir)
    print(os.listdir(dir))
    for author in os.listdir(dir):
      char_num = 0
      words_num = 0
      for file_name in os.listdir(os.path.join(dir, author)):
        with open(os.path.join(dir, author, file_name)) as file:
          text = file.read()
          char_num += len(text)
          words = text.split()
          words_num += len(words)
      print(f'{author}: number of characters: {char_num}, number of words: {words_num}')
      if '_e_' in dir:
        words_str = str(round(words_num/(10**6), 1)).replace('.',',')
        chars_str = str(round(char_num/(10**6), 1)).replace('.',',')
        print(f'${words_str}*10^6$ wyrazów (${chars_str}*10^6$ znaków)')
        print(f'wyrazy: {round((1-words_num/e_basic[author][1])*100, 1)}%')
        print(f'znaki: {round((1-char_num/e_basic[author][0])*100, 1)}%')
      if '_l_' in dir:
        words_str = str(round(words_num/(10**3), 1)).replace('.',',')
        chars_str = str(round(char_num/(10**3), 1)).replace('.',',')
        print(f'${words_str}*10^3$ wyrazów (${chars_str}*10^3$ znaków)')
        print(f'wyrazy: {round((1-words_num/l_basic[author][1])*100, 1)}%')
        print(f'znaki: {round((1-char_num/l_basic[author][0])*100, 1)}%')
    print('\n\n')
```
