```python
from data_preparation import DataPreparation, authors_abbr
from data_fetchers import BookSet
```

### Określenie listy autorów i preprocessingu

```python
preprocessing_list = ['anonymize', 'lower_text','POS_leave_only' ,'lemmatize_text', 'remove_punctuation']#'remove_stop_words', 'remove_punctuation', 'lemmatize_text']#, 'POS_leave_only']
pos = ['NOUN', 'VERB', 'ADJ']
#pos = []
prep_descr = 'set_alomp_NVAdj'
kinds_descr = 'e'
authors = ['Eliza Orzeszkowa', 'Henryk Sienkiewicz', 'Bolesław Prus']
#authors = ['Daniel Naborowski', 'Mikołaj Sęp Szarzyński', 'Elżbieta Drużbacka', 'Adam Mickiewicz', 'Juliusz Słowacki', 'Cyprian Kamil Norwid', 'Adam Asnyk', 'Maria Konopnicka', 'Kazimierz Przerwa-Tetmajer', 'Bolesław Leśmian', 'Jan Kasprowicz']
kinds = ['Epika']
```

```python
bookset = BookSet()
bookset.fetch()

pozytywizm = bookset.list_of_books_to_author_title_list(bookset.get_books_by_authors_list_kinds(authors, kinds))
books_epoch_list = [pozytywizm]
print(books_epoch_list)
```

### Preprocessing

```python
dp = DataPreparation(books_epoch_list)
dp.get_books()
dp.books_preprocessing(preprocessing_list, pos)
```

### Zapisanie zbioru

```python
authors_a = authors_abbr(authors)
dir_path = prep_descr+'_'+kinds_descr+'_'+'_'.join(authors_a)
os.mkdir(dir_path)
```

```python
for author in authors:
  os.mkdir(os.path.join(dir_path, author))
  book_number = 0
  for book in dp.books_text[author]:
    with open(os.path.join(dir_path, author, str(book_number)+'.txt'), 'w') as file:
      file.write(book)
    book_number += 1
```
